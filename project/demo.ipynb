```python
from datasets import load_dataset

label2id = {'OTHER': 0, 'DENTAL_CLICK': 1}
id2label = {v:k for k,v in label2id.items()}

dataset = load_dataset("JBJoyce/DENTAL_CLICK")
```

```python
from IPython.display import Audio
first_click_audio = dataset['test'].filter(lambda example: example['label'] == 1)[0]['audio']['array']
Audio(first_click_audio, rate=16000)
```

```python
from transformers import pipeline
pipe = pipeline(model='JBJoyce/DENTAL_CLICK_classifier')
```

```python
pipe(first_click_audio)
```

```python
dental_clicks_only = dataset['test'].filter(lambda example: example['label'] == 1)
```

```python
for i in range(30): 
    pred = pipe(dental_clicks_only[i]['audio']['array'])[0]
    truth = id2label[dental_clicks_only[i]['label']]
    
    print(f'Truth: {truth}; Prediction: {pred["label"]}.')
```
