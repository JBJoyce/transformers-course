```python
import torchaudio
from IPython.display import Audio, display
from tqdm.auto import tqdm
```

```python
def play_audio(waveform, sample_rate, seconds=60):
  waveform = waveform.numpy()

  num_channels, num_frames = waveform.shape
  if num_channels == 1:
    display(Audio(waveform[0][:seconds * sample_rate], rate=sample_rate))
  elif num_channels == 2:
    display(Audio((waveform[0][:seconds * sample_rate], waveform[1][:seconds * sample_rate]), rate=sample_rate))
  else:
    raise ValueError("Waveform with more than 2 channels are not supported.")
```

```python
metadata = torchaudio.info('/home/jbjoyce/transformers-course/project/audio_files/1.wav')
video_1, sr = torchaudio.load('/home/jbjoyce/transformers-course/project/audio_files/1.wav')

print(metadata)
```

```python
play_audio(video_1, sr, 5)
```

```python
def slice_audio(waveform, sample_rate, segment_duration, directory):
    progress_bar = tqdm(range(0, len(waveform[0]), sample_rate * segment_duration))
    for i, idx in enumerate(range(0, len(waveform[0]), sample_rate * segment_duration)):
        torchaudio.save(
            f'{directory}/video_1_{i}.wav', waveform[:, idx:idx + (sample_rate * segment_duration)], sample_rate,
            encoding="PCM_S", bits_per_sample=16)
        progress_bar.update(1)
        
        
        
```

```python

slice_audio(video_1, sr, 10, '/home/jbjoyce/transformers-course/project/audio_files')
```
